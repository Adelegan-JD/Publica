#### Vectors


**What is a Vector?**

- A vector is simply a list of numbers arranged in order. Think of it like a row of lockers, each holding a value (e.g., age, weight, score, etc)
- Imagine you're holding a shopping list
- ['apples', 'bananas', 'bread'] → text list
- [2, 3, 1] → quantities → This is like a vector!

**Vector Addition**

- When you add vectors, you add each pair of elements
- Bag A = [2, 3]
- Bag B = [1, 5]
- Total items = [2+1, 3+5] = [3, 8]

```python
a = np.array([2, 3])
b = np.array([1, 5])

print("a + b =", a + b)
```

```python
# Vector Addition
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

vector_sum = a + b
print('Vector a:', a)
print('Vector b:', b)
print('Vector sum:', vector_sum)


# Geometric interpretation in 2D
a_2d = np.array([2, 3])
b_2d = np.array([1, 2])
sum_2d = a_2d + b_2d
print('2d Vector A:', a_2d)
print('2d Vector B:', b_2d)
print('2d Vector Sum:', sum_2d)

# plotting figures
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, a_2d[0], a_2d[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)
plt.quiver(0,0, b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005)
plt.quiver(0, 0, sum_2d[0], sum_2d[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005)
plt.quiver(a_2d[0], a_2d[1], b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.003, alpha=0.5)
plt.xlim(-1, 4)
plt.ylim(-1, 6)
plt.grid(True)
plt.title('Vector Addition: Parallelogram Rule')
plt.legend(['Vector A', 'Vector B', 'A + B'])
plt.show()
```

```python
# Properties of vection addition
c = np.array([1, -1, 2])

# Commutative property: a + b = b + a
print(f'a + b = {a + b}' )
print(f'b + a = {b + a}' )
print(f'Commutative property: {np.array_equal(a + b, b + a)}')

# Associative property: (a + b) + c = a + (b + c)
left = (a + b) + c
right = a + (b + c)
print(f'\n(a + b) + c = {left}')
print(f'a + (b + c) = {left}')
print(f'Associative property: {np.array_equal(left, right)}')
```

**Scalar Multiplication**

- Multiply every value in a sector by the same number

- Think of multiplying prices by quantity

```python
prices = np.array([100, 200])
discount = 0.5
print('Half Price:', discount * prices)
```

```python
# Scalar multiplication
vector = np.array([2, 3, 1])
scalar = 3

scaled_vector = scalar * vector
print(f'Original Vector: {vector}')
print(f'Scalar: {scalar}')
print(f'Scaled Vector: {scaled_vector}')

# Geometric interpretation
original_2d = np.array([2, 1])
scales = [0.5, 1, 2, -1]

plt.figure(figsize=(10, 6))
colors = ['red', 'green', 'blue', 'purple']
for i, scale in enumerate(scales):
    scaled = scale * original_2d
    plt.quiver(0, 0, scaled[0], scaled[1], angles='xy', scale_units='xy', scale=1, color=colors[i], width=0.005, label=f'{scale} * v')

plt.xlim(-3, 5)
plt.ylim(-3, 3)
plt.grid(True)
plt.title('Scalar Multiplication Effects')
plt.legend()
plt.show()


```

```python
# Properties of scalar multiplication
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c, c2 = 2, 3

# Associative: c1(c2*a) = (c1*c2)*a
left = c * (c2 * a)
right = (c * c2) * a
print(f'c1(c2*a) = {left}')
print(f'c1(c2*a) = {right}')
print(f'Associative:{np.array_equal(left, right)}')


# Distributive property over vector addition : c(a + b) = ca + cb
left = c * (a + b)
right = (c * a) + (c * b)
print(f'\nc(a + b) ={left}')
print(f'ca + cb = {right}')
print(f'Distributive property: {np.array_equal(left, right)}')
```

**The Physicist's Perspective of a Vector**

Magnitude and Direction 

Vectors have both magnitude(length) and direction in space

```python
import matplotlib.pyplot as plt

# Physicist's point of view: vectors as arrows in space
vector_a = np.array([3, 4])
magnitude = np.linalg.norm(vector_a)
print(f'Vector: {vector_a}')
print(f'Magnitude: {magnitude}')

# Direction (unit vector)
unit_vector = vector_a / magnitude
print(f'Unit Vector (direction): {unit_vector}')

# Visualization
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)
plt.quiver(0, 0, unit_vector[0], unit_vector[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.title('Vector: Magnitude and Direction')
plt.legend(['Original Vector', 'Unit Vector'])
plt.show()

# 
```

**The Mathematician's Pespective**

Abstract Objects
Vectors are any objects that support addition and scalar multiplication

```python
# Mathematician's view: vectors as abstract objects
# Example 1: Polynomials as vectors
# p(x) = 3x^2 + 2x + 1 represented as [1, 2, 3] (coefficient)
poly1 = np.array([1, 2, 3])     # 1 + 2x + 3x^2
poly2 = np.array([2, -1, 1])

# Vector addition for polynomials
poly_sum = poly1 + poly2
print('Polynomial 1:', poly1)
print('Polynomial 2: ', poly2)
print(f'The sum of Polynomial 1 and Polynomial 2 = {poly_sum}')     # (1+2) + (2-1)x + (3+1)x^2 = 3 + x + 4x^2

# Scalar multiplication for polynomials
scalar = 2
poly_scalar = scalar * poly1
print(f'2 * Polynomial 1 =', poly_scalar)

# Example 2: Functions as vectors (function space)
# This is more abstract, but functions can be added and scaled

def func1(x):
    return x**2

def func2(x):
    return 2 * x

# Function addition: (f + g)(x) = f(x) + g(x)
def func_sum(x):
    return func1(x) + func2(x)

x_vals = np.linspace(0, 3, 100)
plt.figure(figsize=(10, 6))
plt.plot(x_vals)
```
